**Question 2:** In how many different states do the customers live in?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"

if not exists(datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!")

unique_states = set()
with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')
  next(csvreader)
  for line in csvreader:
    # Normalize the state string: remove whitespace and convert to uppercase
    state = line[4].strip().upper()
    if state: # Ensure we don't count empty strings
      unique_states.add(state)

print(f"Total unique states (cleaned): {len(unique_states)}")
print(f"States found: {sorted(list(unique_states))}")
```

**Question 3** What is the state with most customers?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
state_counts = {}
# open file and init vars
with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')
  next(csvreader)  # Omitimos el encabezado

  for line in csvreader:
    state = line[4]

    # evaluate the contents of the file
    if state in state_counts:
      state_counts[state] += 1

    else:
      state_counts[state] = 1


most_customer_state = max(state_counts, key=state_counts.get)

print(f"The state with most customer is: {most_customer_state}")
```

**Question 4** What is the state with the least customers?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
state_counts = {}

with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')
  next(csvreader)

  for line in csvreader:
    # Normalize state: strip spaces and convert to uppercase
    state = line[4].strip().upper()
    if state:
      state_counts[state] = state_counts.get(state, 0) + 1

min_count = min(state_counts.values())
least_customer_states = [state for state, count in state_counts.items() if count == min_count]

print(f"Minimum number of customers in a state: {min_count}")
print(f"States with the least customers: {sorted(least_customer_states)}")
```

```python
from os.path import exists
import urllib.request
import csv

url = "https://raw.githubusercontent.com/anyoneai/notebooks/main/customers_and_orders/data/orders.csv"
datafile = "orders.csv"
urllib.request.urlretrieve(url, datafile)

unique_order_count = {}
with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=",")
  next(csvreader)  # Omitimos el encabezado

  for line in csvreader:
    unique_order = line[1]

    if unique_order in unique_order_count:
       unique_order_count[unique_order] += 1

    else:
      unique_order_count[unique_order] = 1


print(f"The number of unique order is: {len(unique_order_count)}")
```

```python
total_items = sum(unique_order_count.values())
total_orders = len(unique_order_count)
average_items_per_order = total_items / total_orders

print(f"The average number of items per order is: {average_items_per_order:.2f}")
```

```python
highest_items_per_order = max(unique_order_count.values())

print(f"The highest number of items per order is: {highest_items_per_order}")
```

```python
import datetime
import csv

october_2021_orders = set()

datafile = "./sample_data/orders.csv" # Explicitly set datafile to orders.csv

with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')
  next(csvreader)  # Skip the header

  for line in csvreader:
    order_id = line[1]
    date_str = line[2]

    # Handle potentially empty date strings or parsing errors
    if date_str.strip(): # Check if the string is not empty after stripping whitespace
      try:
        # Parse the date string into a datetime object
        order_date = datetime.datetime.strptime(date_str.split(' ')[0], '%Y-%m-%d')

        # Check if the order was placed in October 2021
        if order_date.year == 2021 and order_date.month == 10:
          october_2021_orders.add(order_id)
      except ValueError:
        pass # Skip lines with invalid date formats
    else:
      pass # Skip lines with empty date strings

print(f"Number of orders placed in October 2021: {len(october_2021_orders)}")
```

```python
import datetime
import csv

customer_spending_2021 = {}

datafile = "./sample_data/orders.csv" # Explicitly set datafile to orders.csv

with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')
  next(csvreader)  # Skip the header

  for line in csvreader:
    customer_id = line[0]
    date_str = line[2]
    order_total_str = line[3]

    if date_str.strip(): # Check if the date string is not empty
      try:
        order_date = datetime.datetime.strptime(date_str.split(' ')[0], '%Y-%m-%d')

        if order_date.year == 2021:
          try:
            order_total = float(order_total_str)
            customer_spending_2021[customer_id] = customer_spending_2021.get(customer_id, 0.0) + order_total
          except ValueError:
            # Handle cases where order_total_str is not a valid float
            pass # Skip lines with invalid order total formats
      except ValueError:
        pass # Skip lines with invalid date formats

if customer_spending_2021:
  most_spending_customer_id = max(customer_spending_2021, key=customer_spending_2021.get)
  most_spent_amount = customer_spending_2021[most_spending_customer_id]
  print(f"Customer ID {most_spending_customer_id} spent the most in 2021: ${most_spent_amount:.2f}")
else:
  print("No orders found for 2021 or data could not be processed.")
```

```python
import csv

# Load customer names for lookup
customers = {}
with open('./sample_data/customers.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        customers[row[0]] = f"{row[1]} {row[2]}"

# Check specific options provided in the prompt
target_ids = ['5172443', '5415408', '1909369', '5172680', '4700014']

print("Verifying spending for candidates in options:")
for cid in target_ids:
    name = customers.get(cid, "Unknown")
    amount = customer_spending_2021.get(cid, 0.0)
    print(f"ID {cid} ({name}): ${amount:.2f}")

# check the top 5 spenders calculated to see if any names match
top_spenders = sorted(customer_spending_2021.items(), key=lambda x: x[1], reverse=True)[:10]
print("\nGlobal Top 10 Spenders in 2021:")
for cid, amt in top_spenders:
    print(f"{customers.get(cid, 'Unknown')} (ID: {cid}): ${amt:.2f}")
```

```python
import datetime
import csv

monthly_sales = {}

datafile = "./sample_data/orders.csv" # Explicitly set datafile to orders.csv

with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')
  next(csvreader)  # Skip the header

  for line in csvreader:
    date_str = line[2]
    order_total_str = line[3]

    if date_str.strip():
      try:
        order_date = datetime.datetime.strptime(date_str.split(' ')[0], '%Y-%m-%d')
        month_year = order_date.strftime('%Y-%m') # Format as 'YYYY-MM' to distinguish months across years

        try:
          order_total = float(order_total_str)
          monthly_sales[month_year] = monthly_sales.get(month_year, 0.0) + order_total
        except ValueError:
          pass # Skip lines with invalid order total formats
      except ValueError:
        pass # Skip lines with invalid date formats

if monthly_sales:
  best_month_year = max(monthly_sales, key=monthly_sales.get)
  best_month_sales = monthly_sales[best_month_year]
  print(f"The historically best month for sales is {best_month_year} with total sales of: ${best_month_sales:.2f}")
else:
  print("No sales data found or data could not be processed.")
```
